# matplotlib

#### Introduction
- Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

- Matplotlib is open source and we can use it freely. Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

- Matplotlib has many functions as following :
- Style
- Plot
- Scatter
- Pie
- Bar
- Histogram
- Subplot
- Legend

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

## Plot

```python
a = ( 1,2,3,6,5,8,7,4)
plt.plot(a)
```

```python
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18)
plt.plot(a,b)             # a values represent horizental axis and b values represent vertical axis
```

```python
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18)
plt.style.use ('ggplot')   # ggplot is used to show grid
plt.plot(a,b)
```

```python
plt.style.use ('ggplot')
x = np.linspace(0, 10)
y = np.sin(x)     # representing sin function with x values
plt.plot(x, y)
```

```python
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18)
plt.style.use ('ggplot')
plt.ylabel('Time')          # setting axes names
plt.xlabel('Work')
plt.plot(a,b)
```

```python
# using marker
a = (1,2,3,4,5,6)
b = (3,9,10,12,16,18)
plt.style.use ('ggplot')
ax = plt.axes()
ax.set_xlabel('Time')
ax.set_ylabel('Work')
plt.plot(a,b,color='blue',marker='o',markersize=10)
plt.title('time/work consumption')
```

```python
# representing two functions in the same time
def p1(x): return x**4 - 4*x**2 + 3*x
def p2(x): return np.sin(10*x)*10
x = np.linspace(-3,3,1000)
plt.plot(x,p1(x),x,p2(x))
```

```python
# line types : solid,dashed,dashdot and dotted
x=np.arange(0,10,0.005)
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='-')
plt.plot(x, x + 4, linestyle='dashed')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 8, linestyle='dashdot')
plt.plot(x, x + 9, linestyle='-.')
plt.plot(x, x + 12, linestyle='dotted')
plt.plot(x, x + 13, linestyle=':')
```

```python
# line types with colors
x=np.arange(0,10,0.005)
plt.plot(x,x+0,'b--')
plt.plot(x,x+3,'g-')
plt.plot(x,x+6,'r-.')
plt.plot(x,x+9,'y:')
```

```python
# line width
x=np.arange(0,10,0.005)
plt.plot(x,x+0,'b-',linewidth = 1)
plt.plot(x,x+3,'b-',linewidth = 5)
plt.plot(x,x+6,'b-',linewidth = 11)
plt.plot(x,x+9,'b-',linewidth = 20)
```

```python
# line shapes
x=np.arange(0,10,0.5)
plt.plot(x,x,'--')  # dashed
plt.plot(x,x+3,'o') # circles
plt.plot(x,x+6,'^') # triangles
plt.plot(x,x+9,'.') # points
```

```python
# multiple functions in one line
x=np.arange(0,10,0.5)
plt.plot(x,x,'r--',x,x**2,'bo',x,x**3,'g^')
```

## Scatter

```python
x=np.random.normal(0,1,1000)
y=np.random.normal(0,1,1000)
z=np.random.normal(0,1,1000)
w=50*np.random.normal(0,1,1000)
plt.scatter(x,y,c=z,s=w,alpha=0.3)  #alpha represents the transparency,c represents color,s represents size
```

## Pie

```python
plt.pie([15,30,45,10],labels=('Egypt','Syria','Tunise','Morroco'),colors=['red','green','yellow','blue'],autopct ='%1.1f%%')
plt.axis('equal')
plt.show()
```

## Bar

```python
x = np.linspace(0,20,20)
y = np.random.randint(50,size = 20)
plt.bar(x , y,width=0.9,color='r')
plt.text(7, -9, 'Polulation Denisty', style='italic',
bbox={'facecolor':'blue', 'alpha':0.5, 'pad':15}) # nameing of bar with text properties
for xx , yy in zip(x,y):
 plt.text(xx,yy,yy,ha='center',va='bottom')     # putting numbers on bars and adjusting them
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
plt.bar(X, +Y1, color='r',alpha = 0.5)
plt.bar(X, -Y2, facecolor='b',alpha = 0.5)
for x, y in zip(X, Y1):
 plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
 plt.ylim(-1.25, +1.25)
```

## Histogram

```python
from numpy.random import normal
data = np.random.randn(5000)
plt.hist(data, bins=50,alpha=0.5,histtype='step',color='blue')
```

```python
# two histograms
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
for col in 'xy':
 plt.hist(data[col],  alpha=1)
```

```python
# more than one
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
plt.hist(x1,histtype='bar',
alpha=0.3,  bins=100)
plt.hist(x2,histtype='step',
alpha=0.7,  bins=100)
plt.hist(x3,histtype='stepfilled',
alpha=1,  bins=100)
```

## Subplot

```python
x = np.linspace(0,100)
plt.subplot(1,3,1)     # (1,2,1) => (row,column,graph number)
y = 3*x
plt.plot(x,y,c='r')    # graph 1
plt.subplot(1,3,2)
y = -3*x
plt.plot(x,y,c='g')    # graph 2
y=x**2
plt.subplot(1,3,3)
plt.plot(x,y,c='b')    # graph 3

plt.subplots_adjust(left=0,right=1.5
,bottom=0,top=1.5
,wspace=0,hspace=0)
```

```python
for i in range(1, 7):
  plt.subplot(2, 3, i)
  plt.text(0.5, 0.5, str((2, 3, i)) ,fontsize=18, ha='center')
```

## Legend

```python
x = np.linspace(0, 20, 1000)
fig, ax = plt.subplots()
plt.axis('equal')
ax.plot(x, np.sin(x), label='Sin')
ax.plot(x, np.cos(x), label='Cos')
leg = ax.legend()
```

```python
# changing location of the legend
x = np.linspace(0, 20, 1000)
fig, ax = plt.subplots()
plt.axis('equal')
ax.plot(x, np.sin(x), label='Sin')
ax.plot(x, np.cos(x), label='Cos')
ax.legend(loc='upper left', frameon=False)
```
